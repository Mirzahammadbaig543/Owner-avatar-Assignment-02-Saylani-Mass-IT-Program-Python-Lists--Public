1. Write a program that accepts a list from user and print the alternate element of list

```python
# Program to print alternate elements of a list
lst = input("Enter elements of the list separated by spaces: ").split()
print("Alternate elements of the list:")
for i in range(0, len(lst), 2):
    print(lst[i], end=" ")
```

2. Write a program that accepts a list from user. Your program should reverse the content of list and
display it. Do not use reverse() method.

```python
# Program to reverse a list without using the reverse() method
lst = input("Enter elements of the list separated by spaces: ").split()
reversed_lst = lst[::-1]  # Slicing to reverse the list
print("Reversed list:", reversed_lst)
```

3. Find and display the largest number of a list without using built-in function max(). Your program
should ask the user to input values in list from keyboard.

```python
# Program to find the largest number in a list without using max()
lst = input("Enter elements of the list separated by spaces: ").split()
lst = [int(x) for x in lst]  # Convert input strings to integers
largest = lst[0]
for num in lst:
    if num > largest:
        largest = num
print("Largest number in the list:", largest)
```

 4. Write a program that rotates the element of a list so that the element at the first index moves to the
second index, the element in the second index moves to the third index, etc., and the element in the last
index moves to the first index.


```python
# Program to rotate elements of a list
lst = input("Enter elements of the list separated by spaces: ").split()
# Rotate the list
rotated_lst = [lst[-1]] + lst[:-1]
print("Rotated list:", rotated_lst)
```

5. Write a program that input a string and ask user to delete a given word from a string

 6. Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It
should print the date in the form March 12, 2021.

```python
# Program to format a date
from datetime import datetime

date_input = input("Enter a date in the format mm/dd/yyyy: ")
date_obj = datetime.strptime(date_input, "%m/%d/%Y")
formatted_date = date_obj.strftime("%B %d, %Y")
print("Formatted date:", formatted_date)
```

7. Write a program with a function that accepts a string from keyboard and create a new string after
converting character of each word capitalized. For instance, if the sentence is "stop and smell the roses."
the output should be "Stop And Smell The Roses"

```python
# Program to capitalize each word in a string
string = input("Enter a string: ")
capitalized_string = " ".join(word.capitalize() for word in string.split())
print("Capitalized string:", capitalized_string)
```

8. Find the sum of each row of matrix of size m x n. For example for the following matrix output will be
like this :
Sum of row 1 = 32
Sum of row 2 = 31
Sum of row 3 = 63


```python
# Program to find the sum of each row in a matrix
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))
matrix = []

# Input matrix
for i in range(rows):
    row = list(map(int, input(f"Enter row {i+1} elements separated by spaces: ").split()))
    matrix.append(row)

# Calculate and display the sum of each row
for i in range(rows):
    row_sum = sum(matrix[i])
    print(f"Sum of row {i+1} = {row_sum}")
```

9. Write a program to add two matrices of size n x m.

```python
# Program to add two matrices
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

# Input first matrix
print("Enter elements of the first matrix:")
matrix1 = [list(map(int, input().split())) for _ in range(rows)]

# Input second matrix
print("Enter elements of the second matrix:")
matrix2 = [list(map(int, input().split())) for _ in range(rows)]

# Add matrices
result = [[matrix1[i][j] + matrix2[i][j] for j in range(cols)] for i in range(rows)]

# Display result
print("Sum of matrices:")
for row in result:
    print(row)
```

10. Write a program to multiply two matrices

```python
# Program to multiply two matrices
rows1 = int(input("Enter the number of rows for the first matrix: "))
cols1 = int(input("Enter the number of columns for the first matrix: "))
rows2 = int(input("Enter the number of rows for the second matrix: "))
cols2 = int(input("Enter the number of columns for the second matrix: "))

if cols1 != rows2:
    print("Matrix multiplication is not possible!")
else:
    # Input first matrix
    print("Enter elements of the first matrix:")
    matrix1 = [list(map(int, input().split())) for _ in range(rows1)]

    # Input second matrix
    print("Enter elements of the second matrix:")
    matrix2 = [list(map(int, input().split())) for _ in range(rows2)]

    # Multiply matrices
    result = [[0 for _ in range(cols2)] for _ in range(rows1)]
    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    # Display result
    print("Product of matrices:")
    for row in result:
        print(row)
```
